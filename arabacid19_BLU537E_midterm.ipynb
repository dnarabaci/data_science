## Problem 1

```python
def problem1(directory):
    """
    Prints average mileage for each car type. Average of mileages are grouped into columns as New and Used
    
    For each car type finds the difference between maximum and minimum Vehicle Age
    
    Parameters: 

    directory(str): the folder to search for the file
    
    """
    import pandas as pd
 
    #formatting pathing
    from pathlib import Path
    directory = Path(directory)
    
    #read data in
    data=pd.read_csv(directory,encoding="latin-1")
    #create new pivot table with index according to Type attribute and separate according to purchased status
    #and take the mileage mean
    avg_mil=data.pivot_table("Mileage",index="Type",columns="Purchased",aggfunc="mean")
    #display to print nicely
    display(avg_mil)
    #make a new table out of data with the minimum and maximum values columns, then find their difference
    #and assign it to "Vehicle_Age"
    table = pd.pivot_table(data, index=["Type"],aggfunc={"Vehicle_Age": [max, min]})
    #define a function to find the minimum and maximum values and apply it to perform it throughout the axis
    def f(x):
        return pd.Series([x.min(), x.max()], index=['min', 'max'])
    
    table=table.apply(f,axis="columns")
    #make a new column for the difference of min and max and display only that
    table["Vehicle_Age"]=table[min]-table[max]
    display(table[["Vehicle_Age"]])
```

## Problem 2

```python
def problem2(directory):
    """
    Prints the top-10 most frequent Complaint Types
    Prints the number of unique Agency Names
    Prints the top-5 Borough having highest number of "Illegal Parking" Complaints

    For each car type finds the difference between maximum and minimum Vehicle Age

    Parameters: 

    directory(str): the folder to search for the file

    """
    #formatting pathing
    from pathlib import Path
    directory = Path(directory)
    
    import pandas as pd
    #read data in
    data=pd.read_csv(directory,encoding="latin-1")
    #Count the frequencies and assign to display top ten
    freq=data["Complaint Type"].value_counts()
    display(freq[:10])
    print("The number of unique agency names:")
    print(data["Agency Name"].nunique())

    #define a function to sort the illegal parking complaint type, then apply it to grouped dataframe, and count for each borough
    def find(group):
        table=group[group["Complaint Type"]=="Illegal Parking"].sort_values(by="Complaint Type",ascending=False)
        return table

    data1=data.groupby("Complaint Type").apply(find)

    print(data1["Borough"].value_counts())
```

## Problem 3

```python
def problem3(directory):
    """
    Finds the top ten countries having the highest GDP per capita each year and displays it as a table
    
    Parameters: 

    directory(str): the folder to search for the file
    
    """
    #formatting pathing
    from pathlib import Path
    directory = Path(directory)
    
    import pandas as pd
    import numpy as np
    #read in the data as a dataframe, with delimiter as ;
    data=pd.read_csv(directory,encoding="latin-1",delimiter=";")
    #convert with Country name as index
    data=data.pivot_table(index="Country Name")
    #apply lambda function to data and sort each column, then replace each with sorted index, display the top ten
    data=data.apply(lambda f: f.sort_values(ascending=False).index, axis=0).head(n=10)
    #drop the country name column
    data = data.reset_index().drop("Country Name", axis=1)
    #changed the index so it starts at 1 instead of 0 as shown in example
    data.index = np.arange(1, len(data)+1)    
    display(data)
```

## Problem 4

```python
def problem4(directory):
    """
    This function takes data from kaggle, and displays the percentage of churning of clients according to tenure year,
    and also displays the amount of clients in given tenure year.
    
    year_tenure displays the tenure divided by 12 and rounded down.
    Count displays the number of data points
    Percent column shows the percentage of the “Yes” in the churn column
    
    Parameters: 

    directory(str): the folder to search for the file

    """
    #formatting pathing
    from pathlib import Path
    directory = Path(directory)
    
    import pandas as pd
    
    data=pd.read_csv(directory,encoding="latin-1")
    #divide by 12 to get the year values, rounded down
    data["year_tenure"]=data.tenure//12
    #counted all the year_tenure for each tenure, then picked a random column to designate to a new object named Count
    count=data.groupby('year_tenure')["tenure"].count()
    #create a new column for yes or no answers, assign 1 for yes and 0 for no
    data["YesorNo"]=data.Churn.map(lambda x: 1 if "Yes" in x else 0)
    #sum yes values for each tenure year for percentage calculation, assign to a new object
    yes_count=data.groupby("year_tenure")["YesorNo"].sum()
    #merge the data to the main data with count and yes count as year tenure as index
    data=data.merge(count, on="year_tenure")
    data=data.merge(yes_count, on="year_tenure")
    #calculate the yes percentage by dividing the yes or no count to the total count
    data["Percent"]=data["YesorNo_y"]/data["tenure_y"]*100
    #make a new table with index as tenure years, and has the columns percent and total count
    data=data.pivot_table(index="year_tenure")
    data=data[["Percent", "tenure_y"]]
    #rename the column to count for clarity
    data.rename(columns={"tenure_y":"Count"}, inplace=True)
    display(data)
```
